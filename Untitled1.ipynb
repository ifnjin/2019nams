```python
# _*_ coding: UTF-8 _*_
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
```

```python
html = urlopen('http://www.jobkorea.co.kr/recruit/joblist?menucode=cotype4&cotype=8')
bsObject = BeautifulSoup(html,"html.parser")
```

```python
bsObject
```

```python
for link in bsObject.find_all('a'):
    print(link.get_text(), link.get('href'))
```

```python
for td in bsObject.find_all("td", {"class":"tplTit"}):
    title = td.select('a')[0].get_text()
    workplace = td.select('span')[-3].get_text()
    division = td.select('p')[1].get_text()

    print(title,workplace,[division])
```

```python
with open('crawling.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['title', 'work place', 'division'])
        
    for td in bsObject.find_all("td", {"class":"tplTit"}):
        title = td.select('a')[0].get_text()
        workplace = td.select('span')[-3].get_text()
        division = td.select('p')[1].get_text()
        writer.writerow([title,workplace,division])
```
